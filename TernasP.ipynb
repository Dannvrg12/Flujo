El siguiente codigo resuelve el problema de clasificar todas las ternas pitagoricas (x,y,z) tales que x,y,z <=1000

```python
limite = 1000
terna_pitagoricas = []

for x in range(1, limite + 1):
    for y in range(x, limite + 1):
        for z in range(y, limite + 1):
            if x ** 2 + y ** 2 == z ** 2:
                terna_pitagoricas.append((x, y, z))

for terna in terna_pitagoricas:
    print(f"Terna pitagórica: {terna}")
```

Como se puede ver el codigo nos arroja todas las ternas pitagoricas bajo las condiciones dadas en el problema.
